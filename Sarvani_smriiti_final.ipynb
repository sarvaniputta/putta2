```python
import pandas as pd
import bqplot as bq
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import ipywidgets
```

Load the 2017 cps file

```python
cps = pd.read_csv('atuscps_2017.csv')
```

```python
summary = pd.read_csv('atussum_2017.csv') #Summary file of 2017
```

```python
summary.head()
```

```python
cps_ = cps.groupby('TUCASEID').GESTFIPS.first().to_frame()
```

GESTFIPS is the state id which is the fips mapping for the state. bqplot uses fipsmap to identify the states

```python
cps_
```

Merge summary file and cps file by the respondent id


```python
merged = summary.merge(cps_, left_on = 'TUCASEID', right_index = True)
```

```python
merged
```

```python
merged.head()
```

Selected all the 'Relaxing and Leisure' columns. They start with 't1203'

```python
leisure_columns = [x for x in merged.columns if x.startswith('t1203')]

work_columns = [x for x in merged.columns if x.startswith('t0501')]
```

Computed the average time spent for Relaxing and Leisure in each state

```python
leisure_by_state = merged.loc[:, leisure_columns].groupby(merged.GESTFIPS).mean()

work_by_state = merged.loc[:, work_columns].groupby(merged.GESTFIPS).mean()
```

## Map

```python
cscale = bq.ColorScale(scheme = 'viridis')
                       # reverse the colorscale or Hawaii is not visible
                     
map_tt = bq.Tooltip(labels = ['State', 'Time Spent (minutes)'], fields = ['name', 'color'])
sc_geo = bq.AlbersUSA()
states_map = bq.Map(color = leisure_by_state.sum(axis=1).round(1).to_dict(), 
                     map_data=bq.topo_load('map_data/USStatesMap.json'),
                    scales = {'projection':sc_geo, 'color':cscale}, 
                   tooltip = map_tt)

cax = bq.ColorAxis(scale=cscale, orientation='vertical', side='left')
fig = bq.Figure(marks=[states_map],axes=[cax], title = 'Average time spent for leisure')
fig
```

## Histogram

```python
def normalized_hist(data, bins=10):
    counts, bins = np.histogram(data, bins=bins)
    counts = counts*100/counts.sum()
    return bins, counts   
```

```python
sc_x2 = bq.LinearScale()
sc_y2 = bq.LinearScale()
ax_x2 = bq.Axis(label = 'Time spent for leisure', scale=sc_x2, tick_offset = '2000em')
ax_y2 = bq.Axis(label='Frequency%', scale=sc_y2,orientation='vertical')
x, y = normalized_hist(time_spent, bins=14)
bars = bq.Bars(x = x[:-1], y=y, align='left', scales={'x': sc_x2, 'y': sc_y2})
#           ,width=bins[1]-bins[0]  , edgecolor='black')
figure = bq.Figure(marks = [bars],axes=[ax_x2, ax_y2])
figure
```
